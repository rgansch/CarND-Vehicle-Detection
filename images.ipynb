The image loader class was implemented to have a clear interface to the available image sets and to explore the data set. Following categories are available:
- Training vehicle 'train_vehicle': Used to train the classifier on vehicle images
- Training non-vehicle 'train_nonvehicle': Used to train the classifier on non-vehicle images
- Test 'test': used for testing the modules of the pipeline

## Image loading
The used image sets are from the data provided by udacity:
[vehicle](https://s3.amazonaws.com/udacity-sdc/Vehicle_Tracking/vehicles.zip)
[non-vehicle](https://s3.amazonaws.com/udacity-sdc/Vehicle_Tracking/non-vehicles.zip)

```python
%load_ext autoreload
%autoreload 2

from images import ImageLoader
import matplotlib.pyplot as plt

img_loader = ImageLoader()

vehicle_total = img_loader.get_count('train_vehicle')
image_veh,_ = img_loader.get_random('train_vehicle')
plt.imshow(image_veh)
plt.title('Random vehicle (total: %d)' % vehicle_total)
plt.show()

nonvehicle_total = img_loader.get_count('train_non-vehicle')
image_nonveh,_ = img_loader.get_random('train_non-vehicle')
plt.imshow(image_nonveh)
plt.title('Random non-vehicle (total: %d)' % nonvehicle_total)
plt.show()

test_total = img_loader.get_count('test')
image_test,_ = img_loader.get_random('test')
plt.imshow(image_test)
plt.title('Random test (total %d)' % test_total)
plt.show()
```
